```python
import pandas as pd
import pickle
import pylab as pl
import statsmodels.api as sm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
import numpy as np
from statsmodels.distributions.empirical_distribution import ECDF
data = pd.read_csv('data/var12.csv', sep='\s+')
#ЗЧИТУЄМО ЗНАЧЕННЯ З ФАЙЛУ SEP='\S+' РОЗДІЛ ПОМІЖ ЗНАЧЕННЯМИ ЦЕ ПРОБІЛ ДО ЗУСТРІЧІ ІНШОГО ЗНАЧЕННЯ.
df = pd.DataFrame(data)
#ПЕРЕНОСИМО НАШЕ ЗНАЧЕННЯ ТАБЛИЧНУ СТРУКТУРУ ДАННИХ 
print('Mean: {}'.format(df.mean()[0]))
#ВИКОРИСТОВУЄМО ВБУДОВАННУ ФУНКЦІЮ ПОШУКУ СЕРЕДНЬОГО ЗНАЧЕННЯ. ДУЖКИ {}.format 
#ПРОСТО ТИП БІЛЬШ КРАСИВОГО ФОРМАТУВАННЯ 
print('Median: {}'.format(df.median()[0]))
#ВБУДОВАНА ФУНКЦІЯ ПОШУКУ МЕДІАНИ
#[0] БЕРЕМО ТЕЖ ТІЛЬКИ ЗА ДЛЯ ВІДОБРАЖЕННЯ РЕЗУЛЬТАТУ ІНАШКЕ ВИВОДИТЬСЯ ЩЕ ТИП 
# ЗМІННОЇ ТА ДОПИС Value
print('Standart deviation: {}'.format(np.std(df)[0]))
#ВИКОРИСТОВУЄМО ФУНКЦІЮ БІБЛІОТЕКИ ДЛЯ ЗНАХОДЖЕННЯ СТАНДАРТНОГО ВІДХИЛЕННЯ
print('Variance: {}'.format(np.var(df)[0]))
#ВИКОРИСТОВУЄМО ФУНКЦІЮ БІБЛІОТЕКИ ДЛЯ ЗНАХОДЖЕННЯ ДИСПЕРСІЇ
```

```python
v = data['Values']
#ЗАПИСУЄМО КОЛОНКУ Values з таблиці data ЯК ЗМІННУ v
plt.hist(v, color='orange', edgecolor='white', linewidth=1)
#БУДУЄМО ГІСТОГРАММУ ЯКА ПРИЙМАЄ ЗНАЧЕННЯ v
plt.xlabel("Value")
#НАЗВА ОСІ x
plt.ylabel("Frequency")
#НАЗВА ОСІ y
plt.xticks(np.arange(-2, 15, step=1))
#ОБИРАЄМО ЗНАЧЕННЯ ОСІ x ТА КРОК НА ОСІ
plt.yticks(np.arange(0, 26, step=5))
#ОБИРАЄМО ЗНАЧЕННЯ ОСІ y ТА КРОК НА ОСІ
plt.title('Values in variant 12')
#НАЗВА ДІАГРАММИ
plt.show()
#ВИВОДИМО ДІАГРАММУ
```
